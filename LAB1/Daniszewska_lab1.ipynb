```python
import numpy as np
import matplotlib.pyplot as plt

# Define the size N
N = 9  # Length of the xµ vector

# Define the xµ vector for variant 4
x_mu = np.array([6, 2, 4, 3, 4, 5, 0, 0, 0], dtype=complex)

# Create matrix K
k = np.arange(N).reshape(N, 1)  # Row and column vectors
mu = np.arange(N).reshape(1, N)
K = k @ mu  # Outer product

# Create Fourier matrix W
W = np.exp(1j * 2 * np.pi * K / N)

# Compute IDFT using matrix notation
x_k = (1/N) * W @ x_mu

# Plot the synthesized signal
plt.figure(figsize=(8, 5))
plt.stem(np.real(x_k))  # Removed use_line_collection for compatibility
plt.title('Synthesized signal (IDFT)')
plt.xlabel('Index k')
plt.ylabel('Amplitude')
plt.grid()
plt.show()

# Display matrices W and K
print("Matrix K:")
print(K)
print("\nMatrix W:")
print(W)

x_k_check = np.fft.ifft(x_mu)
print("Verification result:", np.allclose(x_k, x_k_check))

print("Max absolute difference:", np.max(np.abs(x_k - x_k_check)))
```
